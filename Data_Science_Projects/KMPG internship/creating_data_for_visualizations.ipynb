```python
import pandas as pd
```

```python
df = pd.read_csv('kmpg_insternship_cleaned_data.csv',index_col='transaction_id').sort_index()
df.head()
```

```python
df.columns
```

```python
df['age']=pd.DatetimeIndex(df.transaction_date).year-pd.DatetimeIndex(df.DOB).year
df.age
```

```python
for x in df.age[df.age>120].index:
    df.drop(x, inplace=True)
```

```python
df.drop('DOB',axis=1, inplace=True)
```

```python
df.to_csv('kmpg_virtual_interm.csv')
```
